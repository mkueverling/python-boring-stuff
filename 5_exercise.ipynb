#### TIPPS:
1. Das angegebene Beispiel muss nur passend erweitert werden.
2. Denke daran, dass nun 3 Werte in einer angezeigt werden müssen. Abstände etc. sind zu ermitteln

`````python
import math # Mathematische-Funktionen

## Sinuskurve zeichnen
f = 20              # Gibt eine Frequenz an (Sinnvolle ganzzahlige Werte liegen zwischen 1 und 50)
MaxWert = 1.        # Der maximale "Ausschlag" der Funktion positiv wie negativ
Skalierung = 40.    # Meint, dass für den Bereich -1 bis 0, 0 bis 1 usw. jeweils 40 Zeichen verwendet werden.

i = 0               # Zähler
while i<100:
    w = math.sin(i*f/100.)                  # Wert
    Leer = int(w * Skalierung + Skalierung) # Anzahl auszugebender Leerstellen berechnen   
    print(round(w,3),"\t"," "*Leer,'+')
    i = i+1

print("Ende")
`````

### PI über Kugel ermitteln (12P)
Schreibe ein Programm, bei dem mittels Monte-Carlo-Methode PI aufgrund des Zusammenhangs bei einer Kugel (Volumen = 4/3 PI r^3) ermittelt wird.

TIPP: Schreibe das unten angegebene Programm passend um.


`````python
import random   # Hole die Bibliothek für Zufallszahlen
 
def estimate_pi(n_points):  # Schätze PI ab
    points_inside_circle = 0
    total_points = 0
 
    while total_points<n_points:
        x = random.random() # Zufallszahl zwischen 0 und 1
        y = random.random()
        distance = x**2 + y**2
        if distance <= 1:
            points_inside_circle += 1
        total_points += 1

    return 4 * points_inside_circle / total_points

print(estimate_pi(100)) # Abschätzung von PI
`````
