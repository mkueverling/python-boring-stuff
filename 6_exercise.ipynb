## 6.2 Spiel "Nicoma"
Wir möchten eine äußerst alte Form eines arithmetischen Rätsels programmieren. In "Arithmetica" von Nicomachus (+ ca. 120 v.Chr.) wirst Du aufgefordert, Dir eine Zahl zwischen 1 und 100 auszudenken. Teile diese Zahl durch 3, 5 und 7 und teile jedesmal den Rest mit. Der schlaue Computer Nicoma "errät" dann die von Dir ausgedachte Zahl.

Folgender Dialog soll entstehen (in ".." sind die Antworten)

````
Denke Dir eine Zahl zwischen 1 und 100 aus.
Welcher Rest verbleibt, wenn Du die Zahl durch 3 teilst? "1"
Welcher Rest verbleibt, wenn Du die Zahl durch 5 teilst? "1"
Welcher Rest verbleibt, wenn Du die Zahl durch 7 teilst? "1"

Ich muss überlegen... Die Zahl ist 1.

Noch einmal? (J oder N) "J"

Denke Dir eine Zahl zwischen 1 und 100 aus.
Welcher Rest verbleibt, wenn Du die Zahl durch 3 teilst?

usw... bis auf die Frage *Noch einmal? (J oder N)* "N" eingegeben wird.
````

Hinweis: Wenn die drei Eingaben gespeichert werden in den Variablen `r3`, `r5` und `r7`, dann kannst Du die Zahl raten lassen durch 
1. `zahl=70*r3+21*r5+15*r7`
2. `zahl>105 dann zahl=zahl-105`
3. Prüfedoch, ob das tatsächlich immer passt! (Test-Programm mit Schleife)
4. Könnte die korrekt zu "ratende" Zahl auch größer sein? Wenn ja, wie groß? Und warum?

## 6.3 "Nimm"-Spiel

Das "Nimm"-Spiel ist ein Spiel für zwei Personen.

Gegeben ist ein Haufen von 13 Streichhölzern. Abwechselnd nehmen zwei Personen jeweils 1, 2 oder 3 Streichhölzer. Derjenige, der den Haufen leerräumt, hat gewonnen.

Erstellen Sie einen Algorithmus für das oben beschriebene Einstiegsproblem (z.B. PAP oder Struktogramm). Es hilft, wenn man das Spiel selbst einmal durchspielt.

Implementieren Sie diesen Algorithmus in einem Programm, sodass zwei Spieler gegeneinander spielen können.

### Mögliche Erweiterungen:
1. Erweitern Sie die Aufgabe, sodass die Anzahl der Starthölzer und die Anzahl der jeweils maximal wegzunehmenden Hölzer eingegeben werden kann. Das Programm soll fehlerhafte (Zahl-)Eingabe melden.
2. Erweitere das Programm, sodass es gegen einen Spieler spielt. Das Programm spielt in einer ersten Version zufällig.
3. Jetzt versucht das Programm überlegt mitspielen. (Schwierig!)
