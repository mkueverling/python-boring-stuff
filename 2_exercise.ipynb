## 2.2 Notenausgabe
Schreiben Sie eine Funktion `Note2Str()`, die eine Note gemäß der Prüfungsordnung in eine Zeichenkette umwandelt und diese zurückgibt.
* Input: Eine mögliche Note (1, 1.3, 1.7, 2.0,..., 3.7, 4, 5)
* Output: Die ausgeschriebene Note
    * 1.3 -> sehr gut (minus)
    * 2.7 -> befriedigend (plus)
    * 4 -> Ausreichend

* Kann man das evtl. schlau aufteilen? Entwickeln Sie Algorithmus, der möglichst wenige `if..else` benötigt.

### Beispiele
````
print(Note2Str(2))      # gut
print(Note2Str(3.3))    # befriedigend (minus)
print(Note2Str(7))      # ungültige Note (7)
````

`````python
def Note2Str(note):
    haupt_note = {
        1:'Sehr gut',
        2:'Gut',
        3:'Befriedidgend',
        4:'Ausreichend',
        5:'Mangelhaft'}

    sub_note = {
        0.0:'',
        0.3:' minus',
        -0.3:' plus'}
    
    basis = round(note)
    diff = round(note - basis, 1)

    if basis in haupt_note and diff in sub_note:
        string_note = haupt_note[basis] + sub_note[diff]
        return string_note 
    else:
        return f'falscher input'


    
Note2Str(2.7)

`````

### Tipps
* Schreiben Sie eine Funktion `TageBisOstern()`. Hier verwenden Sie `TageImJahr()`.
* Schreiben Sie dann eine Funktion `DauerSession()`, die die Funktion `TageBisOstern()` verwendet.

`````python
def berechne_ostern(jahr):
    a = jahr % 19
    b = jahr % 4
    c = jahr % 7
    k = jahr // 100
    p = (8 * k + 13) // 25
    q = k // 4
    m = (15 + k - p - q) % 30
    d = (19 * a + m) % 30
    n = (4 + k - q) % 7
    e = (2 * b + 4 * c + 6 * d + n) % 7

    tag = 22 + d + e
    monat = 3

    if tag > 31:
        tag = tag - 31
        monat = 4
    
    return tag, monat
`````

`````python
def tag_nummer(tag, monat, jahr):
    # 1. jan - 28. feb
    tage_bis_maerz = 59
    

    if (jahr % 400 == 0) or (jahr % 4 == 0 and jahr % 100 != 0):
        # 1. jan - 29. feb
        tage_bis_maerz = 60 
        
    # Jetzt rechnen wir die laufende Nummer des Tages aus
    if monat == 3:
        return tage_bis_maerz + tag
    elif monat == 4:
        return tage_bis_maerz + 31 + tag 
`````

`````python
def dauer_session(start_jahr):
    end_jahr = start_jahr + 1
    
    # finde ostertag
    oster_tag, oster_monat = berechne_ostern(end_jahr)
    
    # der wievielte tag im jahr ist ostertag
    oster_nummer = tag_nummer(oster_tag, oster_monat, end_jahr)
    
    # aschermitwoch 46 tage vor ostern
    aschermittwoch_nummer = oster_nummer - 46
    
    # rest vom alten jahr (51 ist fixed vom 11.11.) + tage im neuen jahr
    # -1 aschermittwoch
    gesamtdauer = 51 + aschermittwoch_nummer
    
    return gesamtdauer

print(f"Session 2025/2026: {dauer_session(2025)} Tage")
`````
