## 2.2 Notenausgabe
Schreiben Sie eine Funktion `Note2Str()`, die eine Note gemäß der Prüfungsordnung in eine Zeichenkette umwandelt und diese zurückgibt.
* Input: Eine mögliche Note (1, 1.3, 1.7, 2.0,..., 3.7, 4, 5)
* Output: Die ausgeschriebene Note
    * 1.3 -> sehr gut (minus)
    * 2.7 -> befriedigend (plus)
    * 4 -> Ausreichend

* Kann man das evtl. schlau aufteilen? Entwickeln Sie Algorithmus, der möglichst wenige `if..else` benötigt.

### Beispiele
````
print(Note2Str(2))      # gut
print(Note2Str(3.3))    # befriedigend (minus)
print(Note2Str(7))      # ungültige Note (7)
````
