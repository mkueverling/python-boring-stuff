## Treibstoffverbrauch
Sie sind 60 km gefahren und mussten 2,6 Liter Diesel tanken. Ein Liter Diesel kostet 1,64.

1. Definieren Sie für diese Werte drei Variablen
2. Berechne in einer zusätzlichen Variable den Verbrauch in l pro 100 km und gib den Wert aus.
3. Berechne in einer zusätzlichen Variable die Kosten pro km und gib den Wert aus.

```python
# 1
km = 60
tank = 2.6
preis_liter = 1.64

verbrauch_100km =(tank/km)*100
kosten_pro_km = ((verbrauch_100km / 100) * preis_liter)

print(f'Verbrauch pro 100km liegt bei {verbrauch_100km:.2f} Liter')
print(f'Preis pro km liegt bei {kosten_pro_km:.2f} Euro')

```

## Quadratische Gleichung
1. Suche die Lösung für eine quadratische Gleichung der Form x²+px+q=0
2. Verwende gültige Werte wie p=5 und q=-8, 
2. Schreibe ein Programm zur Lösung der Gleichung.

Tipp: Sie kommen ohne math.sqrt(...) aus! `**` ist der Potenzoperator 

```python
# Lösungen einer quadratischen Gleichung mit der p/q-Formel 

p = 5
q = -8

term_unter_wurzel = (p / 2)**2 - q

wurzel = term_unter_wurzel ** 0.5

x1 = -(p / 2) + wurzel
x2 = -(p / 2) - wurzel

print(f'x1 = {x1} und x2 = {x2}')
```
