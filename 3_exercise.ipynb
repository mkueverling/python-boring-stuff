## 3.2 Schleifen
Es sollen folgende Zahlenreihen ausgegeben werden (innerhalb der Schleife darf nur eine Ausgabeanweisung stehen!):
Für jede Zahlenreihe schreiben Sie eine Funktion `printReihe1(n)`, die die Ausgabe der Reihe innerhalb der Funktion übernimmt.

Testen Sie Ihre Funktionen jeweils mit n=0, n=10, n=100 und n=-5

1. : 0, 1, 2, 3, …, n
2. : -n, -8, -6, …, +n
3. : 0, 1, 4, 9, … bis zum maximalen Wert `n`
4. : 1, 2, 4, 8, … bis zum maximalen Wert `n`
5. : 0, 1, 3, 6, 10, 15, … bis zum maximalen Wert `n`

`````python
# 1
print('Reihe 2: ')

def printReihe1(n):
    print('\nReihe 1: ')
    for x in range(n + 1):      
        print(x, end=' ')

        
printReihe1(100)

def printReihe2(n):
    print('\nReihe 2: ')
    for x in range(-n, n + 1):
        print(x, end=' ')

printReihe2(10)

def printReihe3(n):
    print('\nReihe 3: ')
    i = 0
    while i * i <= n:
        print(i * i, end=' ')
        i += 1
         
printReihe3(100)

def printReihe4(n):
    print('\nReihe 4: ')
    i = 1
    while i <= n:
        print(i, end=' ')
        i = i * 2

printReihe4(100)

def printReihe5(n):
    print('\nReihe 5: ')
    i = 0
    j = 1
    while i <= n:
        print(i, end=" ")
        i = i+j
        j = j + 1

printReihe5(100)

`````
